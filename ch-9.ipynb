```python
class grandparent:
    def gp (self):
        print("This is Grand Parent.")
class Parent (grandparent):
    def f1 (self):
        print("This is parent class.")
class Child (Parent):
    def f2 (self):
        print("This is child class.")
ob = Parent()
ob.f2()
ob.f1()
ob.gp()
```
