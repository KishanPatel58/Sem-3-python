```python
# multiple inheritance.
class parent1:
    def f2 (self):
        print("This is parent 1")
class parent2:
    def f1 (self):
        print("This is parent 2")
class child (parent1,parent2):
    def f (self):
        print("This is child")
ob = child()
ob.f()
ob.f1()
ob.f2()
```

```python
# PB (633) :-
'''
Implement the following hierarchy . The Staff function has name and salary as its data members, 
the derived class Teaching has subject as its data member and the class NonTeaching has 
department as its data member. The derived class method overrides (extends) the methods of the base class.
'''
class Staff:
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
    def dis(self):
        print("Name :",self.name)
        print("Salary :",self.salary)
class Teaching (Staff):
    def __init__(self, name, salary, subject):
        print("-: Teaching :-")
        super().__init__(name,salary)
        self.subject = subject
    def dis (self):
        super().dis()
        print("Subject :",self.subject)
class NonTeaching (Staff):
    def __init__(self, name, salary, sub):
        print("-: Non Teaching :-")
        super().__init__(name, salary)
        self.sub = sub
    def dis (self):
        super().dis()
        print("Subject :",self.sub)
ob = Teaching("A",50000,"JAVA")
ob.dis()
ob1 = NonTeaching("B",30000,"Digital Marketing")
ob1.dis()
```

#### 2. Comperision Operator :-
##### __lt__(<)

##### __ge__(>=)
##### __eq__(==)
##### __ne__(!=)

```python
# Comparision Operator
class Cl:
    def __init__(self,x):
        self.x = x
    def __lt__(self,other):
        if(self.x < other.x):
            print(other.x,"is Big")
        else:
            print(other.x,"is Small")
    def __gt__(self,other):
        if(self.x > other.x):
            print(other.x,"is Small")
        else:
            print(other.x,"is Big")
    def __le__(self,other):
        if(self.x <= other.x):
            print(other.x,"is Big or Equal")
        else:
            print(other.x,"is Small")
    def __ge__(self,other):
        if(self.x >= other.x):
            print(other.x,"is Less or Equal")
        else:
            print(other.x,"is Big")
    def __eq__(self,other):
        if(self.x == other.x):
            print(other.x,"Both Equal")
        else:
            print(other.x,"Not Equal")
    def __ne__(self,other):
        if(self.x != other.x):
            print(other.x,"Not Equal")
        else:
            print(other.x,"Both Equal")
ob = Cl(10)
ob1 = Cl(20)
ob>ob1
ob==ob1
ob<ob1
```

```python
# Arithmetic Operator
class Cl:
    def __init__(self,x):
        self.x = x
    def __add__(self,other):
        return self.x + other.x
    def __sub__(self,other):
        return self.x - other.x
    def __mul__(self,other):
        return self.x * other.x
    def __truediv__(self,other):
        return self.x / other.x
    def __floordiv__(self,other):
        return self.x // other.x
    def __mod__(self,other):
        return self.x % other.x
    def __pow__(self,other):
        return self.x ** other.x
ob = Cl(10)
ob1 = Cl(20)
print("Ans is: ",ob*ob1)
```
