## Scatter plot

```python
from numpy import random
import matplotlib.pyplot as plt
x = random.randint(100,size=200)
y = random.randint(50,size=200)

plt.scatter(x,y)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
x = random.randint(100,size=200)
y = random.randint(50,size=200)

plt.scatter(x,y, s=20, c="red", alpha=0.8)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
x = random.randint(100,size=200)
y = random.randint(50,size=200)
size = random.randint(80,size=200)

plt.scatter(x,y, s=size, c="red", alpha=0.8)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
x = random.randint(100,size=200)
y = random.randint(50,size=200)
size = random.randint(80,size=200)
color = random.randint(80,size=200)

plt.scatter(x,y, s=size, c=color, alpha=0.8)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
x = random.randint(100,size=200)
y = random.randint(50,size=200)
size = random.randint(80,size=200)
color = random.randint(80,size=200)

plt.scatter(x,y, s=size, c=color, alpha=0.5)
plt.colorbar()
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
x = random.randint(100,size=150)
y = random.randint(50,size=150)
size = random.randint(100,size=150)
color = random.randint(100,size=150)

plt.scatter(x,y, s=size, c=color, alpha=0.8, cmap="plasma")
plt.xlabel("X - Scale")
plt.ylabel("Y - Scale")
plt.colorbar()
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
x = random.randint(100,size=150)
y = random.randint(50,size=150)
x1 = random.randint(50,size=200)
y1 = random.randint(50,size=200)

plt.scatter(x,y, s=10, c="black", alpha=1)
plt.scatter(x1,y1, s=10, c="red", alpha=1)
plt.colorbar()
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
x = random.randint(100,size=150)
y = random.randint(50,size=150)
x1 = random.randint(50,size=200)
y1 = random.randint(50,size=200)

plt.subplot(1,2,1)
plt.scatter(x,y, s=10, c="black", alpha=1)
plt.subplot(1,2,2)
plt.scatter(x1,y1, s=10, c="red", alpha=1)
plt.colorbar()
plt.show()
```
