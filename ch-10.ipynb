```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y1 = [10,5,15,3,18]
y2 = [5,10,3,15,20]
plt.plot(x,y1, marker="o", ms=15, mfc="pink",mec="r",color="blueviolet", ls="-.", lw=5)
plt.plot(x,y2, marker="+", ms=15, mfc="pink",mec="r",color="green", ls=":", lw=2)
plt.show()
```
