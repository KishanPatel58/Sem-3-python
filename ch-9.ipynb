```python
class grandparent:
    def gp (self):
        print("This is Grand Parent.")
class Parent (grandparent):
    def f1 (self):
        print("This is parent class.")
class Child (Parent):
    def f2 (self):
        print("This is child class.")
ob = Parent()
ob.f2()
ob.f1()
ob.gp()
```

```python
# multiple inheritance.
class parent1:
    def f2 (self):
        print("This is parent 1")
class parent2:
    def f1 (self):
        print("This is parent 2")
class child (parent1,parent2):
    def f (self):
        print("This is child")
ob = child()
ob.f()
ob.f1()
ob.f2()
```

```python
# over riding.
class parent1:
    def f2 (self):
        print("This is parent 1")
class parent2:
    def f (self):
        print("This is parent 2")
class child (parent1,parent2):
    def f (self):
        super().f()
        print("This is child")
ob = child()
ob.f()
ob.f2()
```

```python
# over riding.
class parent1:
    def f2 (self):
        print("This is parent 1")
class parent2:
    def f (self):
        print("This is parent 2")
class child (parent1,parent2):
    def f (self):
        parent2.f(self)
        print("This is child")
ob = child()
ob.f()
ob.f2()
```

```python
# MRO (method resolution order)
class A:
    def t1 (self):
        print("This is Class A")
class B (A):
    def t1 (self):
        print("This is Class B")
class C:
    def t1 (self):
        print("This is Class C")
class D (B,C):
    def t (self):
        print("This is Class D")
ob = D()
ob.t()
ob.t1()
print(D.__mro__)
```

```python
class A:
    def f1 (self):
        print("class A")
class B:
    def f1 (self):
        print("class B")
class C (A,B):
    def f1 (self):
        super().f1()
        B.f1(self)
        print("Class C")
ob = C()
ob.f1()
```

```python
# Hierarchical inheritance.
class A:
    def f1 (self):
        print("class A")
class B (A):
    def c1 (self):
        print("child 1")
class C (A):
    def c2 (self):
        print("child 2")
ob = B()
ob.c1()
ob.f1()

ok = C()
ok.c2()
ok.f1()

print(issubclass(B,A))
print(isinstance(ob,A))
```
